## Restoring the Image using ContraHarmonic self coded

```python
from math import floor, ceil
myImage = Image.open('noisyFinalImage.jpg').convert('L')
```

```python
def applyingFilter(input_img, filter,M,N):
    
    n, m = len(filter), len(filter[0])  # m is height, n is width
    a, b = m / 2, n / 2  # size of neighborhood
    a = int(a)
    b = int(b)

    # get transpose of the 1-d filter
    if isinstance(filter, np.ndarray):
        wt = filter.ravel()
    else:
        wt = np.array(filter).ravel()

    def correlate(x, y):
        z = np.zeros(n * m)  # pad with zeros
        z = np.full(n * m, input_img[x, y])  # pad with border duplicates
        # fill in available neighborhood
        for i in range(x - a, x + a + 1):
            for j in range(y - b, y + b + 1):
                if i >= 0 and i < (M -1) and j >= 0 and j < (N-1):
                    print(z[(i - x + a) * n + j - y + b], M,N)
                    z[(i - x + a) * n + j - y + b] = input_img[i, j]
        return np.dot(wt, z)
    # apply to each pixel
    xx, yy = np.meshgrid(range(M), range(N), indexing='ij')
    vf = np.vectorize(correlate)
    return vf(xx, yy)    
```

```python
def power(var, power):
    return np.power(var, power) if power>0 else np.power(1/var, power)


def contraHarmonicMean(myImage,Q,size):
    M,N = myImage.size
    imageArray =np.asarray(myImage,dtype=np.float64)
    numerator = power(imageArray,Q+1 )
    denominator = power(imageArray,Q)
    kernel = np.full(size,1.0)
    result = applyingFilter(numerator, kernel,M,N) / applyingFilter(denominator, kernel,M,N)
    contraHarmonicPic = Image.fromarray(result)
    return (contraHarmonicPic)
```

```python
image = cv2.imread('noisyFinalImage.jpg',1)
myImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
def power(var, power):
    return np.power(var, power) if power>0 else np.power(1/var, power)

windowSize = 3
Q= -1
kernelSize = windowSize - 2
imgArray2 = np.copy(myImage)
for i in range(kernelSize,myImage.shape[0]-kernelSize):
    for j in range(kernelSize,myImage.shape[1]-kernelSize):
        neumerator = 0
        denominator = 0
        kernel = myImage[i-kernelSize:i+kernelSize+1, j-kernelSize:j+kernelSize+1].flatten()
        for k in kernel:
            neumerator = neumerator + power(k,Q+1)
            denominator = denominator + power(k,Q)
        result = int(np.ceil(np.divide(neumerator,denominator)))
        imgArray2[i][j] = result
contraImage = Image.fromarray(imgArray2)
plt.imshow(contraImage , cmap = matplotlib.cm.Greys_r) 
plt.show()
```

## Restoring the noisy image using Midpoint Filter. 

Logic behind the Code: 
Apply min filter to the Image to get the minimum of the pixels.
Apply max filter to the image to get the maximum of the pixels. 
Divide it by 2 and take the average to get midpoint,


### Code required for midpoint

```python
def midpoint(imageArray, size=1):
    newImage = np.copy(imageArray)
    for i in range(size,imageArray.shape[0]-size):
        for j in range(size,imageArray.shape[1]-size):
            numerator = 0
            denominator = 0
            kernel = imageArray[i-size:i+size+1, j-size:j+size+1]
            minimum = int(np.min(kernel))
            maximum = int(np.max(kernel))
            numerator = (minimum+maximum)
            denominator = 2
            answer = np.divide(numerator,denominator)
            newImage[i][j] = int(answer)
    midPointPic = Image.fromarray(newImage)
    return(midPointPic)
```

### Restoring the Image with both Noises Added

```python
myImage = Image.open('noisyFinalImage.jpg').convert('L')
imageArray = np.asarray(myImage)
result = midpoint(imageArray)
plt.imshow(result , 'gray') 
plt.show()
```
