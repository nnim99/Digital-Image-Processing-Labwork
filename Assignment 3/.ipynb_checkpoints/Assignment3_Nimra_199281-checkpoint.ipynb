## Restoring the Image using ContraHarmonic self coded

### Restoring Both noise added Image

```python
from math import floor, ceil
myImage = Image.open('noisyFinalImage.jpg').convert('L')
```

```python
image = cv2.imread('noisyFinalImage.jpg',1)
myImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
def power(var, power):
    return np.power(var, power) if power>0 else np.power(1/var, power)

wSize = 3
Q= 1
kernelSize = wSize - 2
imageArray = np.copy(myImage)
for i in range(kernelSize,myImage.shape[0]-kernelSize):
    for j in range(kernelSize,myImage.shape[1]-kernelSize):
        numerator = 0
        denominator = 0
        kernel = myImage[i-kernelSize:i+kernelSize+1, j-kernelSize:j+kernelSize+1].flatten()
        for k in kernel:
            numerator = numerator + power(k,Q+1)
            denominator = denominator + power(k,Q)
        result = int(np.ceil(np.divide(numerator,denominator)))
        imageArray[i][j] = result
contraImage = Image.fromarray(imageArray)
plt.imshow(contraImage , cmap = matplotlib.cm.Greys_r) 
plt.show()
```

## Restoring the noisy image using Midpoint Filter. 

Logic behind the Code: 
Apply min filter to the Image to get the minimum of the pixels.
Apply max filter to the image to get the maximum of the pixels. 
Divide it by 2 and take the average to get midpoint,


### Code required for midpoint

```python
def midpoint(imageArray, size=1):
    newImage = np.copy(imageArray)
    for i in range(size,imageArray.shape[0]-size):
        for j in range(size,imageArray.shape[1]-size):
            numerator = 0
            denominator = 0
            kernel = imageArray[i-size:i+size+1, j-size:j+size+1]
            minimum = int(np.min(kernel))
            maximum = int(np.max(kernel))
            numerator = (minimum+maximum)
            denominator = 2
            answer = np.divide(numerator,denominator)
            newImage[i][j] = int(answer)
    midPointPic = Image.fromarray(newImage)
    return(midPointPic)
```

### Restoring the Image with both Noises Added

```python
myImage = Image.open('noisyFinalImage.jpg').convert('L')
imageArray = np.asarray(myImage)
result = midpoint(imageArray)
plt.imshow(result , 'gray') 
plt.show()
```

### Restoring the Image with Gaussian Noise only.

## Observations

In the above restoration, when average filtering was used. It showed good results for both Gaussian and Noise. However it did not remove salt and pepper as good as it removed gaussian noise. 

When we used contra harmonic with positive value, to gaussian noise it enhanced the image well. whereas as the value was positive for Q. It only eliminated pepper noise well. Salt noise is still present. 

For Median filter, It worked extremely well for the salt and pepper noise, almost the image was fully recovered. It was also good for gaussian. Overall 90% of noise was eliminated.

Midpoint filter, was not good for salt and pepper as the value can lie in between of noise, and can easily distort the image. However it worked really well for the Gaussian Noise as observed above.
