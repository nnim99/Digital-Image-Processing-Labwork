## Adding Guassian noise

######  For adding Gaussian Noise, I am using np.random.normal inbuilt function which is used to generate Noise. I am keeping variance as 36 so that sigma can be 6 as it gives visible result for gaussian noise

```python
mean = 0  
variance = 36
sigma = variance **0.5
gaussNoise = np.random.normal(mean, sigma, myImage.shape)
gaussNoise = gaussNoise.reshape(myImage.shape)
noisyImage = myImage + gaussNoise

plt.imshow(noisyImage,'gray')
plt.show()
cv2.imwrite('noisyImage.jpg',noisyImage)
```

## Restoring the Image using ContraHarmonic self coded

```python
from math import floor, ceil
image = cv2.imread('noisyFinalImage.jpg',1)
myImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
def contraHarmonicMean(myImage,Q,kSize):
    M,N = myImage.shape
    imageArray =np.asarray(myImage)
    for i in range(kSize,myImage.shape[0]-kSize):
        for j in range(kSize,myImage.shape[1]-kSize):
            numerator = 0
            denominator = 0
            kernel = myImage[i-kSize:i+kSize+1, j-kSize:j+kSize+1].flatten()
            for k in kernel:
                numerator = numerator + np.power(k,(Q+1))
                denominator = denominator + np.power(k,(Q))
                print(denominator)
            result = int(np.ceil(np.divide(numerator,denominator)))
            imageArray[i][j] = result
    contraHarmonicPic = Image.fromarray(imageArray)
    
    return (contraHarmonicPic)
```

```python
print(myImage.shape)
result = contraHarmonicMean(myImage,-1,1)
display(result)
```

## Restoring the noisy image using Midpoint Filter. 

Logic behind the Code: 
Apply min filter to the Image to get the minimum of the pixels.
Apply max filter to the image to get the maximum of the pixels. 
Divide it by 2 and take the average to get midpoint,


### Code required for midpoint

```python
def midpoint(imageArray, size=1):
    newImage = np.copy(imageArray)
    for i in range(size,imageArray.shape[0]-size):
        for j in range(size,imageArray.shape[1]-size):
            numerator = 0
            denominator = 0
            kernel = imageArray[i-size:i+size+1, j-size:j+size+1]
            minimum = int(np.min(kernel))
            maximum = int(np.max(kernel))
            numerator = (minimum+maximum)
            denominator = 2
            answer = np.divide(numerator,denominator)
            newImage[i][j] = int(answer)
    midPointPic = Image.fromarray(newImage)
    return(midPointPic)
```

### Restoring the Image with Gaussian Noise only.

```python
myImage = Image.open('noisyImage.jpg').convert('L')
imageArray = np.asarray(myImage)
result = midpoint(imageArray)
plt.imshow(result , 'gray') 
plt.show()
```

### Restoring the Image with Salt and Pepper Noise only.

```python
myImage = Image.open('SPImage.jpg').convert('L')
imageArray = np.asarray(myImage)
result = midpoint(imageArray)
plt.imshow(result , 'gray') 
plt.show()
```
