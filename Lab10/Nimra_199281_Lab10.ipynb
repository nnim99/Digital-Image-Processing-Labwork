# Task 1

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('task1.png')
kernal = np.ones((5,5),np.uint8)

ret,thresh1 = cv2.threshold(image,10,255,cv2.THRESH_BINARY)
plt.imshow(thresh1,'gray')
plt.show()
```

```python
dilation = cv2.dilate(thresh1,kernal,iterations=1)
cv2.imwrite('result1.png',dilation)
plt.imshow(dilation,'gray')
plt.show()
```

```python
erosion = cv2.erode(thresh1,kernal,iterations = 1)
cv2.imwrite("result2.png", erosion)
plt.imshow(erosion,'gray')
plt.show()
```
