```python
import numpy as np
import cv2
import PIL
from PIL import Image
from matplotlib import pyplot as plt
import matplotlib
import matplotlib.pyplot as plt
```

# Negating function

```python
def invertingAllImage(image):
    if (image.mode == "RGB"):
        width, height = image.size
        arrayImage = np.array(image)
        for i in range(height):
            for j in range(width):
                red = abs(arrayImage[i][j][0] - 255)
                green = abs(arrayImage[i][j][1] - 255)
                blue = abs(arrayImage[i][j][2] - 255)
                arrayImage[i][j]=[red,green,blue]
        image01 = Image.fromarray(arrayImage)
    elif (image.mode == "L"):
        width, height = image.size
        arrayImage = np.array(image)
        for i in range(height):
            for j in range(width):
                
                pixel = abs(arrayImage[i][j] - 255)
                
                arrayImage[i][j]=pixel
        image01 = Image.fromarray(arrayImage)
    return(image01)
    
```

```python
image = Image.open('lena_color.gif')
```

# RGB to negative

```python
rgb_image = image.convert('RGB')
rgb_image.save('lena_color.png')
print(image.size)
plt.imshow(rgb_image)
plt.show()
print(rgb_image.mode)
negateImage = invertingAllImage(rgb_image)
plt.imshow(negateImage)
plt.show()
```

# Grayscale Invert

```python
gray_image = image.convert('L')
plt.imshow(gray_image,'gray')
plt.show()
print(gray_image.mode)
negateImage = invertingAllImage(gray_image)
plt.imshow(negateImage,'gray')
plt.show()
```

# Inverting black and white

```python
image = cv2.imread('lena_color.png')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, bw_img = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
plt.imshow(bw_img,'gray')

plt.show()
cv2.imwrite('bw.png',bw_img)
bw_img = Image.open('bw.png')
negateImage = invertingAllImage(bw_img)
plt.imshow(negateImage,'gray')
plt.show()
```

## Horizontal gradient

```python
image = Image.open('lena_color.png').convert('L')
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#grayArray = np.array(gray_image)
width, height = image.size
gray_image = image.load()

for x in range(width-1):
    for y in range(height):
        gray_image[x,y] = gray_image[ x+1,y] - gray_image[x,y]
image.save('new.png')
new= Image.open('new.png')
plt.imshow(new,'gray')
plt.show()
```

## Vertical gradient 

```python
image = Image.open('lena_color.png').convert('L')
#gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#grayArray = np.array(gray_image)
width, height = image.size
gray_image = image.load()

for x in range(width):
    for y in range(height-1):
        gray_image[x,y] = gray_image[ x,y+1]- gray_image[x,y]
image.save('new.png')
new= Image.open('new.png')
plt.imshow(new,'gray')
plt.show()
```
