```python
from skimage import io, color
import matplotlib.pyplot as plt
import numpy as np
from skimage import exposure
import pylab

def convolving(image, kernel,length):

    
    kernel = np.flipud(np.fliplr(kernel))    # Flip the kernel
    output = np.zeros_like(image)            # convolution output
    # Add zero padding to the input image
    pad=int(length/2)
    image_padded = np.zeros((image.shape[0] + length-1, image.shape[1] + length-1))   
    image_padded[pad:-pad, pad:-pad] = image
    print(image_padded)
    for x in range(image.shape[1]):     # Loop over every pixel of the image
        for y in range(image.shape[0]):
            output[y,x]=(kernel*image_padded[y:y+length,x:x+length]).sum()        
    return output
```

```python
kernel1 = np.ones((3,3),np.float32)/9
kernel2 = np.ones((5,5),np.float32)/25
kernel3 = np.ones((15,15),np.float32)/225


kernel4 = np.array((
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1]), np.float32)/16
```

```python

image = cv2.imread('smoothing.tif')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
result = convolving(gray, kernel1,3)
plt.imshow(gray, 'gray')
plt.show()
plt.imshow(result,'gray')
plt.show()
cv2.imwrite('task11.png',result)
```

```python
unsharp = cv2.imread('unsharpmasking.tif')    
unsharp = cv2.cvtColor(unsharp, cv2.COLOR_BGR2GRAY)

gaussian = np.array([[1,1,2,2,2,1,1],
                     [1,2,2,4,2,2,1],
                     [2,2,4,8,4,2,2],
                     [2,4,8,16,8,4,2],
                     [2,2,4,8,4,2,2],
                     [1,2,2,4,2,2,1],
                     [1,1,2,2,2,1,1]])/140

gaussianFilter = convolving(unsharp,gaussian,7)


plt.imshow(gaussianFilter, cmap=plt.cm.gray)
plt.axis('off')
plt.show()

unsharped = unsharp - gaussianFilter
result=Image.fromarray(unsharped).save("task3.png")


plt.imshow(unsharped, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
kernel15 = np.ones((15,15))/225

result = convolving(unsharp,kernel15,15)

plt.imshow(result, cmap=plt.cm.gray)
plt.axis('off')
plt.show()

unsharped = unsharp - result
cv2.imwrite("task32.png",unsharped)

plt.imshow(unsharped, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image

image= Image.open('saltandpaper.tif')
A= np.asarray(image)
pad= np.pad(A, (1, 1), 'constant')
display(image)

fil= np.ones((3,3),dtype=np.uint8)

smooth = np.ones(A.shape, dtype = np.uint8)

for row in range(len(pad)-2):
    for col in range(len(pad[0])-2):
                     med = 0
                     for r in range(len(fil)):
                         for c in range(len(fil[0])):
                             fil[r][c]*= pad[row+r][col+c]
                     med= np.median(np.sort(fil,axis=None))
                     smooth[row][col]= med
                     fil= np.ones((3,3),dtype=np.uint8)

smoothimage=Image.fromarray(smooth)
```
